```python
import os
```

```python
word = "miroro" + ".wav"
base_path = "/audio dialect"

os.path.join(base_path, word)
```

```python
dialect = df["DIALECT"]
```

```python
vezo = []
```

```python
def create_dict(words, base_path):
    words_json = []
    for index, word in enumerate(words):
        word = word + ".wav"
        path = os.path.join(base_path, word)
        if os.path.exists(path):
            mot = {
                "id": index,
                "dialect": word[index],
                "path": path,
            }
            words_json.append(mot)
    return words_json
```

```python
for index, word in enumerate(dialect):
    word = word + ".wav"
    base_path = "audio_dialect"
    path = os.path.join(base_path, word)
    if os.path.exists(path):
        mot = {
            "id": index,
            "dialect": dialect[index],
            "path": path,
        }
        vezo.append(mot)
```

```python
vezo
```

```python
import json
```

```python
json_data = json.dumps(vezo, indent=4)

print(json_data)

# Sauvegarder dans un fichier
with open("vezo.json", "w") as json_file:
    json_file.write(json_data)
```
